```python
import pandas as pd
```

```python
import os

dataset_folder_path = r"io/legal_test"  # Use r for raw string to handle backslashes
dataset = [os.path.join(dataset_folder_path, entry.name) for entry in os.scandir(dataset_folder_path) if entry.is_file()]

train_folder_path = r"io/legal_train"  # Use r for raw string to handle backslashes
train = [os.path.join(train_folder_path, entry.name) for entry in os.scandir(train_folder_path) if entry.is_file()]
```

```python
import json

test = {
    "question": [],
    "query": [],
    "category": []
}


for file in dataset:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    for item in data:
        test["question"].append(item["question"])
        test["query"].append(item["query"])
        test["category"].append(file)
```

```python
df_test = pd.DataFrame(test)
df_test
```

```python
df_test = df_test.drop_duplicates(subset="query")
df_test
```

```python
import json

train_set = {
    "question": [],
    "query": [],
    "category": []
}

for file in train:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    for item in data:
        train_set["question"].append(item["question"])
        train_set["query"].append(item["query"])
        train_set["category"].append(file)
```

```python
df_train = pd.DataFrame(train_set)
df_train
```

```python
df_train = df_train.drop_duplicates(subset="query")
df_train
```
